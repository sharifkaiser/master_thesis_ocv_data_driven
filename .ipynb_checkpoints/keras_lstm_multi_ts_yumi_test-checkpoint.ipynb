## Highlight of notebook:
1. how to fit a LSTM model to predict a point in time series given another time series.
2. how to extract weights for forget gates, input gates and output gates from the LSTM's model
3. how to calculate hidden and cell states from the weight outputs.


```python

import matplotlib.pyplot as plt
import tensorflow as tf
from keras.backend import set_session
import keras
import seaborn as sns
import pandas as pd
import sys, time
import numpy as np
import warnings
warnings.filterwarnings("ignore")
print("python {}".format(sys.version))
print("keras version {}".format(keras.__version__))
print("tensorflow version {}".format(tf.__version__))
config = tf.compat.v1.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.95
config.gpu_options.visible_device_list = "1"
set_session(tf.compat.v1.Session(config=config))

def set_seed(sd=123):
    from numpy.random import seed
    import random as rn
    ## numpy random seed
    seed(sd)
    ## core python's random number 
    rn.seed(sd)
    ## tensor flow's random number
#     set_random_seed(sd)
    tf.random.set_seed(sd)

```

```python
# Create a synthetic time series data. Having a data in front help understood the meaning of each parameter in LSTM

def random_sample(len_ts=3000,D=1001):
    c_range = range(5,100)
    c1 = np.random.choice(c_range)
    u = np.random.random(1)
    const = -1.0/len_ts
    ts = np.arange(0,len_ts)
    
    x1 = np.cos(ts/float(1.0 + c1))
    x1 = x1*ts*u*const
    
    y1 = np.zeros(len_ts)

    for t in range(D,len_ts):
        ## the output time series depend on input as follows: 
        y1[t] = x1[t-2]*x1[t-D] 
    y = np.array([y1]).T
    X = np.array([x1]).T
    return y, X
def generate_data(D= 1001,Nsequence = 1000,T=4000, seed=123):
    X_train = []
    y_train = []
    set_seed(sd=seed)
    for isequence in range(Nsequence):
        y, X = random_sample(T,D=D)
        X_train.append(X)
        y_train.append(y)
    return np.array(X_train),np.array(y_train)

D = 10      # seq length
T = 1000    # 1000 points
X, y = generate_data(D=D,T=T,Nsequence = 100)    #Nsequence means number of total time series. Here, we 100 different time serieses, each time series has 1000 time points
print(X.shape, y.shape)
```

#### The data explanation:
1. There are total k time series, e.g. k=100 (in code, it is Nsequence=100)
2. For each time series $X_t(t=1,2, ..,T) $, there is a lagged version $Y_t$ as output (dependent) time series, T = 1000 e.g., each time series has total 1000 data points or time points
3. The lag is given by $D$ in code, $D = 10$ means dependent each time series $Y_t$ lags 10 values from $X_t$
4. We generated 100 independent sets of time series (as mentioned in #1)
5. As there is a lag of 10, first 10 of each $Y_t$ are zero (no value)

```python
print(X.shape[0]) # k =100, means 100 independent time series
print(X.shape[1]) # T =1000, means 1000 data points for each time series
# D = Time lag, output ts y is 10 lagged from input ts X
# The first 10 time samples are set as zero
print(y[0:10])
```

```python
# plot 16 diff time series input and output
def plot_examples(X,y,ypreds=None,nm_ypreds=None):
    fig = plt.figure(figsize=(16,10))
    fig.subplots_adjust(hspace = 0.32,wspace = 0.15)
    count = 1
    n_ts = 16
    for irow in range(n_ts):
        ax = fig.add_subplot(int(n_ts/4),4,count)
        ax.set_ylim(-0.5,0.5)
        ax.plot(X[irow,:,0],"--",label="x1")
        ax.plot(y[irow,:,:],label="y",linewidth=3,alpha = 0.5)
        ax.set_title("{:}th time series sample".format(irow))
        if ypreds is not None:
            for ypred,nm in zip(ypreds,nm_ypreds):
                ax.plot(ypred[irow,:,:],label=nm)   
        count += 1
    plt.legend()
    plt.show()
plot_examples(X,y,ypreds=None,nm_ypreds=None)
```

### LSTM model (LSTM single layer)

1. LSTM model also have hidden states that are updated between recurrent cells (time). In fact, the LSTM layer has two types of states: hidden state and cell states that are passed between the LSTM cells. However, only hidden states are passed to the next layer
2. input gate, forget gate, output gate, cell state, hidden state
3. LSTM params = hunits * hunits + hunits * nfeats + hunits * 1   [hunits = hidden units, nfeats = # of features of the time series]. Notice, the first one is always hunits, because this is just the number of units we use. The next things are: hunits for $h_{t-1}$, nfeats for $x_t$ and 1 for bias 
5. $h_t$ is passed as input to next layer e.g., dense or fully connected layer for regression
6. Each of the gates are thought as a standard neuron in a feed forward neural network

Equations: notice, $x_t$ (nfeats e.g., #features) and $h_{t-1}$ (hidden from prev time stamp, hunits is defined by us) in all equations. These 4 equations give computation for #params in #3 above. Think of U, h, W etc. as matrix or vectors. Each weight W, U always have dimension of hunits, x is #features and bias b is always 1 value. Sigma, tanh etc. are all elementwise, all  <br>
Input gate $i_t=sigmoid(U_ih_{t−1}+W_ix_t+b_i)$ <br>
Forget gate $f_t=sigmoid(U_fh_{t−1}+W_fx_t+b_f)$<br>
New candidate cell state $\tilde{c}=tanh(U_ch_{t−1}+W_cx_t+b_c)$<br>
Output gate $o_t=sigmoid(U_oh_{t−1}+W_ox_t+b_o)$<br>

Rest of the equations:<br>
$c_t = f_t * c_{t-1} + i_t * \tilde{c} $<br>
$h_t = o_t * tanh(c_{t}) $<br>

```python
from keras import models 
from keras import layers

def define_model(len_ts,
                 hidden_neurons = 1,
                 nfeature=1,
                 batch_size=None,
                 stateful=False):
    in_out_neurons = 1
    
    inp = layers.Input(batch_shape= (batch_size, len_ts, nfeature),
                       name="input")  

    rnn = layers.LSTM(hidden_neurons, 
                    return_sequences=True,
                    stateful=stateful,
                    name="RNN")(inp)

    dens = layers.TimeDistributed(layers.Dense(in_out_neurons,name="dense"))(rnn)
    model = models.Model(inputs=[inp],outputs=[dens])
    
    model.compile(loss="mean_squared_error",
                  sample_weight_mode="temporal",
                  optimizer="rmsprop")
    return(model,(inp,rnn,dens))
```

```python
X_train, y_train = X ,y
hunits = 3
model1, _ = define_model(
    hidden_neurons = hunits,
    len_ts = X_train.shape[1])   # length of TS is 1000
model1.summary()

# notice, total params 4 cells *(3*3 + 3*1 + 3*1) = 60
```

```python
# init weight matrix, As the first 10 Yt is not defined based on Xt, not compute the MSE for the first 10 samples (hence, set zero weight) and compute it only for the t = 10,...,1000.

w = np.zeros(y_train.shape[:2])
w[:,D:] = 1    # first 10 of each ts is zero, rest is 1 because first 10 does not
w_train = w
print(y_train.shape[:2])   # output is (100, 1000) = (k, T)
print(w[0])
```

```python
from keras.callbacks import ModelCheckpoint
start = time.time()
hist1 = model1.fit(X_train, y_train, 
                   batch_size=2**9,   # batch_size 512
                   epochs=200, 
                   verbose=2,
                   sample_weight=w_train,
                   validation_split=0.05,
                   callbacks=[
                   ModelCheckpoint(filepath="weights{epoch:03d}.hdf5")])
end = time.time()
print("Time took {:3.1f} min".format((end-start)/60))
```

```python
labels = ["loss","val_loss"]
for lab in labels:
    plt.plot(hist1.history[lab],label=lab + " model1")
plt.yscale("log")
plt.legend()
plt.show()
# Notice the training loss is above the validation loss i.e., overfitting. This is because the data is not enough only 100 individual time series (k=100). As I produce more data, e.g., k=1000,
# it solves the overfitting problem 
```
